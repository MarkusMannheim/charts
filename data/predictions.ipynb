```python
import pandas as pd
import numpy as np

data = pd.read_csv("./predictions.csv", parse_dates=["month"])

for i in data.index:
    if pd.isna(data.at[i, "current"]):
        prediction = data.at[i, "prediction"]
        data.at[i, "lower"] = prediction - 3 / np.sqrt(2)
        data.at[i, "upper"] = prediction + 3 / np.sqrt(2)

data.to_csv("./predictions.csv", index=False)
data
```
